```python
import pandas as pd
data = {
    "name": ["Asha", "Sanjana", "Rahul", "Nimish"],
    "age": [23, 24, 25, 24],
    "salary": [30000, 35000, 25000, 28000],
    "department": ["HR", "IT", "Finance", "Marketing"]
}
df = pd.DataFrame(data)
print("dataframe: \n", df)
print("Average Salary:\n", df["salary"].mean())
print("salary >= 30000: \n", df[df["salary"]>=30000])
print("sorted salary: \n", df.sort_values("salary"))
print("Mean of the salary: \n", df["salary"].median)
```
